```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/content/Churn_Modelling.csv")
x = dataset.iloc[:,3:13].values
y = dataset.iloc[:,13].values
```

```python
dataset.head(5)
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer # Corrected import path
import numpy as np

labelencoder_x_1 = LabelEncoder()
x[:,1] = labelencoder_x_1.fit_transform(x[:,1])
labelencoder_x_2 = LabelEncoder()
x[:,2] = labelencoder_x_2.fit_transform(x[:,2])

# Explicitly convert the array to a numeric type (float) after all encoding
x = x.astype(float)

# Handle any remaining NaN values in x using SimpleImputer before splitting/scaling
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
x = imputer.fit_transform(x)

x
```

```python
# spliting the dataset info the training set and test set
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2, random_state=0)
```

```python
len(dataset)
```

```python
len(x_test)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
# classical Ml
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
pred = clf.predict(x_test)
```

```python
#  making confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test ,pred)
print(cm)
```

```python
#  making the confusion matrix
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test,pred)
print(score*100) # this 100 to show percentage
```

Accuracy from Decision TREE classifier


```python
# Random forest
# classical Ml

from sklearn.ensemble import RandomForestClassifier
clf2 = RandomForestClassifier(n_estimators=200)
clf2.fit(x_train,y_train)
pred2 = clf2.predict(x_test)
```

```python
# making the confusion matrix
from sklearn.metrics import confusion_matrix
cm2 = confusion_matrix(y_test , pred2)
print(cm2)
```

```python
# checking accuracy
from sklearn.metrics import accuracy_score
score2 = accuracy_score(y_test,pred2)
print(score2*100)
```

**Accuracy of Random Forest classifier is 86%**

**ANN**

```python
# importing the keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
import keras
from keras.models import Sequential
from keras.layers import Dense, Input

classifier = Sequential()
# Adding the input layer and the first hidden layer
classifier.add(Input(shape=(10,))) # Explicitly defining the input layer
classifier.add(Dense(units=6,kernel_initializer ='uniform',activation='relu'))
# Adding the second hidden layer
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
# Adding the third layer
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
# Adding the output layer
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
len(x_train)
```

```python
# compiling the ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
y_train[0]
# fitting the ANN to the training set
classifier.fit(x_train,y_train,batch_size=10,epochs=50)
```

```python
# Making prediction and evaluating the model
# predicting the test set result
y_pred_ann = classifier.predict(x_test)
y_pred_ann = (y_pred_ann > 0.5)

cm3 = confusion_matrix(y_test,y_pred_ann)
print(cm3)
```

```python
from sklearn.metrics import accuracy_score
score3 = accuracy_score(y_test,y_pred_ann)
print(score3*100)
```

**Accuracy from ANN:84%**
