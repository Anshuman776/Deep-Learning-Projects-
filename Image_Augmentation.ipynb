```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import random
from skimage import exposure
from skimage.util import random_noise
from skimage import transform
from cv2 import resize
```

```python
img = mpimg.imread('/content/cat.avif')
```

```python
plt.imshow(img)
```

**# Rescaling**

```python
img_rescale = resize(img, (150,150))
plt.imshow(img_rescale)
```

this image is blurry image as we can see from previous image

```python
img_rescale = resize(img, (50,50))
plt.imshow(img_rescale)
```

**# Flipping**

```python
horizontal_flip = np.fliplr(img)
plt.imshow(horizontal_flip)
```

```python
Vertical_flip = np.flipud(img)
plt.imshow(Vertical_flip)
```

**# Rotation**

```python
from skimage import transform
tran_img = transform.rotate(img,random.uniform(-40,40))
plt.imshow(tran_img)
```

**# Adding Noise**

```python
img_noise = random_noise(img , mode ='s&p',clip = True)
plt.imshow(img_noise)
```
