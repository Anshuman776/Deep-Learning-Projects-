```python
from keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input , decode_predictions
```

here vgg16 us pre trained model

```python
# load the vgg16 pre-trained modelos and specify the input shape
model = VGG16(weights='imagenet', include_top=True,input_shape=(224,224,3))
```

```python
import numpy as np
img_path ="/content/mic.webp"
img = image.load_img(img_path,target_size=(224,224))
x = image.img_to_array(img)
x=np.expand_dims(x,axis=0)
x=preprocess_input(x)
```

```python
# use the model to predict the task
pred = model.predict(x)
```

```python
pred
```

in this when i see on goole its shows that there are 1000 classes and 92% accuracy

```python
# get the top 5 predictions with class names
decode_pred = decode_predictions(pred,top=5)[0]
```

```python
for p in decode_pred:
  print(p[1],":",p[2])
```
